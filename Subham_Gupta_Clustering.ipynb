```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering, Birch
from sklearn.mixture import GaussianMixture
from sklearn.metrics import davies_bouldin_score
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict

# Load datasets
customers = pd.read_csv("Customers.csv")
products = pd.read_csv("Products.csv")
transactions = pd.read_csv("Transactions.csv")

# Merge datasets
data = transactions.merge(customers, on="CustomerID").merge(products, on="ProductID")

# Feature engineering
data['TransactionDate'] = pd.to_datetime(data['TransactionDate'])
data['SignupDate'] = pd.to_datetime(data['SignupDate'])
data['days_active'] = (data['TransactionDate'].max() - data['SignupDate']).dt.days

# Aggregated customer-level features
customer_features = data.groupby('CustomerID').agg(
    total_transactions=('TransactionID', 'count'),
    total_value=('TotalValue', 'sum'),
    avg_value=('TotalValue', 'mean'),
    product_variety=('ProductID', 'nunique'),
    days_active=('days_active', 'max'),
).reset_index()

# One-hot encode 'Region' column
region_encoded = pd.get_dummies(customers['Region'], prefix='Region')
customer_features = pd.concat([customer_features, region_encoded], axis=1)

# Handle missing values
customer_features = customer_features.dropna()

# Normalize features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(customer_features.drop(columns=['CustomerID']))

# Clustering algorithms
algorithms = {
    "K-Means": KMeans,
    "Agglomerative": AgglomerativeClustering,
    "DBSCAN": DBSCAN,
    "GMM": GaussianMixture,
    "Birch": Birch,
    "Spectral": SpectralClustering
}

# Define cluster ranges and store DBI scores
cluster_ranges = range(2, 11)
db_scores = defaultdict(list)

# Loop over clustering algorithms
for name, algorithm in algorithms.items():
    for n_clusters in cluster_ranges if name not in ["DBSCAN"] else [None]:  # DBSCAN doesn't use n_clusters
        # Initialize the model with algorithm-specific parameters
        if name == "K-Means":
            model = algorithm(n_clusters=n_clusters, random_state=42, init='k-means++')
        elif name == "Agglomerative":
            model = algorithm(n_clusters=n_clusters, linkage='ward')
        elif name == "DBSCAN":
            model = algorithm(eps=1.5, min_samples=5)
        elif name == "GMM":
            model = algorithm(n_components=n_clusters, covariance_type='full', random_state=42)
        elif name == "Birch":
            model = algorithm(n_clusters=n_clusters, threshold=0.5, branching_factor=50)
        elif name == "Spectral":
            model = algorithm(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)
        
        # Fit model and predict clusters
        if name == "GMM":
            model.fit(scaled_features)
            clusters = model.predict(scaled_features)
        else:
            clusters = model.fit_predict(scaled_features)
        
        # Calculate Davies-Bouldin Index
        if len(set(clusters)) > 1:  # Ensure DBI is calculable
            db_index = davies_bouldin_score(scaled_features, clusters)
            db_scores[name].append((n_clusters, db_index))
            
            # Visualize clusters using PCA
            pca = PCA(n_components=2)
            reduced_features = pca.fit_transform(scaled_features)
            
            plt.figure(figsize=(10, 6))
            sns.scatterplot(
                x=reduced_features[:, 0], y=reduced_features[:, 1],
                hue=clusters, palette='viridis', style=clusters
            )
            plt.title(f"{name} Clusters (K={n_clusters})")
            plt.xlabel("PCA Component 1")
            plt.ylabel("PCA Component 2")
            plt.legend(title="Cluster")
            plt.show()
        
        # Dendrogram for Agglomerative Clustering
        if name == "Agglomerative":
            linked = linkage(scaled_features, method='ward')
            plt.figure(figsize=(10, 6))
            dendrogram(linked, truncate_mode='lastp', p=n_clusters, show_leaf_counts=True)
            plt.title(f"Dendrogram for {name} Clusters (K={n_clusters})")
            plt.xlabel("Samples")
            plt.ylabel("Distance")
            plt.show()

# Display DB Index values for each algorithm
for name, scores in db_scores.items():
    print(f"{name} DB Index values:")
    for n_clusters, db_index in scores:
        print(f"  Number of Clusters: {n_clusters}, DB Index: {db_index:.4f}")

# Plot DB Index comparison
plt.figure(figsize=(12, 6))
for name, scores in db_scores.items():
    x, y = zip(*scores)
    plt.plot(x, y, marker='o', label=name)
plt.title("DB Index vs Number of Clusters for Different Algorithms")
plt.xlabel("Number of Clusters")
plt.ylabel("Davies-Bouldin Index")
plt.legend()
plt.grid(True)
plt.show()
```
